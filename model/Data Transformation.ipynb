```python
import pandas as pd
from pathlib import Path
```

```python
#Read csv
malicious_csv = pd.read_csv('../data/malicious_only.csv')
benign_csv = pd.read_csv('../data/benign_only.csv')
all_csv = malicious_csv.append(benign_csv, ignore_index = True)
```

```python
#Rename malicous to malicious
all_csv.rename(columns={"malicous":"malicious"}, inplace=True)
```

```python
#Checking unique domain name
all_csv['domain_name'].nunique()
```

```python
#Checking unique ip
all_csv['ip'].nunique()
```

```python
#Factorizing domain_name
code, uniques = all_csv['domain_name'].factorize()
all_csv['domain_name'] = code

#Factorizing ips
code, uniques = all_csv['ip'].factorize()
all_csv['ip'] = code

#Factorizing countries
code, uniques = all_csv['country'].factorize()
all_csv['country'] = code

#Deleting url column
all_csv.drop(['url'], axis=1, inplace=True)


all_csv.head()
```

```python
all_csv.tail()
```

```python
#Converting to csv
filepath = Path('../data/complete_transformed_CSV.csv')
all_csv.to_csv(filepath)
```
