```python
import pandas as pd
from pathlib import Path
```

```python
#Read csv
all_csv = pd.read_csv('../test_data/featured_data.csv')
print(all_csv)
```

```python
#Deleting url column

benign = all_csv.loc[all_csv["type"].isin(["benign"])]
benign = benign.sample(n=200000)
all_csv.drop(all_csv.index[(all_csv["type"].isin(["benign"]))],axis=0,inplace=True)
all_csv = pd.concat([all_csv, benign])

code, uniques = all_csv['tld'].factorize()
all_csv['tld'] = code

code, uniques = all_csv['type'].factorize()
all_csv['type'] = code
print(all_csv['type'].value_counts())
```

```python
#Converting to csv
filepath = Path('../test_data/ready_for_training.csv')
all_csv.to_csv(filepath, index=False)
```
